```python
import lightkurve as lk
import pandas as pd
```

```python


# Search and download all files
search = lk.search_lightcurve("EPIC 246199087", mission="K2", campaign=12)
lc_collection = search.download_all()

# Stitch them together
lc = lc_collection.stitch()

# Clean and process
lc = lc.remove_nans().remove_outliers()
lc = lc.flatten(window_length=301)
lc = lc.normalize()

```

```python
# Convert to DataFrame
df = lc.to_pandas()


# Save to CSV for Excel
df.to_csv("trappist1e_lightcurve.csv")
```
