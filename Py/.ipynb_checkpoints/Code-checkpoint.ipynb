```python
from lightkurve import read
import matplotlib.pyplot as plt
import numpy as np

lc = read("ktwo246199087-c12_llc.fits")
# lc.plot()
```

```python

# Remove bad quality
lc = lc[lc.quality == 0]

# Normalize
lc = lc.normalize()

lc.scatter()
plt.show()
```

```python
lc = lc[(lc.flux > 0.95) & (lc.flux < 1.05)]
lc.scatter()
plt.show()
```

```python
dip = lc[(lc.time.value > 2928) & (lc.time.value < 2930)]
dip.scatter()
plt.show()

depth = 1 - dip.flux.min()
print("Transit depth:", depth)
```
